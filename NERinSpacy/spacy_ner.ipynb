```python
import json
import spacy
import random
```

```python
with open('C:/Users/naikp/Downloads/annotations1.json') as fp: #loading training data
    training_data = json.load(fp)   ####dictionary
```

```python
training_data
```

```python
# prepare an empty model to train
nlp = spacy.blank('en')
# nlp = spacy.load('en_core_web_sm')
nlp.vocab.vectors.name = 'demo'
ner = nlp.create_pipe('ner')
nlp.add_pipe(ner, last = True)
```

```python
# Add the custome NER Tags as entities into the model
for label in training_data["classes"]:
    print(label)
    nlp.entity.add_label(label)
```

```python
# Train the model
optimizer = nlp.begin_training()
```

```python
import random
for iter in range(100):
    for text, annotations in training_data["annotations"]:
#         random.shuffle(training_data)
        if len(text) > 0:
            print([text])
            nlp.update([text], [annotations], sgd=optimizer)
```
