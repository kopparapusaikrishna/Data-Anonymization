```python
columns = list(df_join1)
print(columns)
```

```python
from tqdm import tqdm
```

```python
data = []

for i in tqdm(range(df_join1.shape[0])):
    lst = []
    for j in range(df_join1.shape[1]):
        s = df_join1[columns[j]][i]
        lst.append(s[1:(len(s)-1)])
    lst1 = []
    lst1.append(lst)
    data.append(lst1)
```

```python
print(data[0:5])
```

```python
input_df = pd.DataFrame(data, columns=['tokens'])
```

```python
def lslicing(lst):
    l = []
    for i in lst:
        l.append(i[1:-1])
    return l
```

```python
input_list = []
for i in tqdm(range(len(input1_df["tokens"]))):
    string = input1_df["tokens"][i]
    res = string.strip('][').split(', ')
    res1 = lslicing(res)
    input_list.append(res1)
```

```python
input_list[0]
```

## Word2Vec Model

```python
model = gensim.models.Word2Vec(min_count=1,
                               vector_size= 50,
                               workers=4,
                               window =3,
                               sg = 1)
```

```python
model.build_vocab(input_list, progress_per=10000)
```

```python
len(model.wv.key_to_index)
```

```python
model.wv.key_to_index
```

```python
model.train(input_list, total_examples=model.corpus_count, epochs=model.epochs)
```
